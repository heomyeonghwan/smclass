```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

데이터 수정

```python
## replace : 데이터값을변경 , inplace=True 지정
df['학교'].replace({'구로고':'가산고'},inplace=True)
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)

## 특정 컬럼 값 변경
# 컬럼 =  변경값을 입력
df['SW특기'] = df['SW특기'].str.lower()
df['SW특기'] = df['SW특기'].str.upper()
df['SW특기'].str.capitalize()  #capitalize() e첫글자 대문자
df['키'] =200

# 컬럼명 = 문자 + 문자
df['학교'] = df['학교']+'등학교'

# 특정셀 지정 변경
df.loc['1번','국어'] =100
df.loc['2번',['국어','영어']] = 10
```

데이터 추가

-컬럼추가: df[컬럼명]

-row 추가 : df.loc[index명]

```python
## 데이터 추가
## 컬럼에 없는 이름을 입력하면 컬럼이 추가가 됨
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계'] / 5
df
```

조건식을 사용해서 값변경

```python
df['결과'] ='Fail' 

# 결과 컬럼에서 조건식을 사용해서 값을변경
filt = df['합계']>=400
df.loc[filt,'결과'] ='pass'
df
```

```python

df['합산'] = 100
## 국어점수가 90점 이상이면 합산에 +100
filt = df['국어'] >=90
df.loc[filt,'합산'] = df['합산']+100
df
```

drop 삭제  - columns

```python
# # 데이터 삭제 - drop : columns, 적용 : inplace=True)
# df.drop(columns=['합산'],inplace=True)
# # 컬럼2개삭제
df.drop(columns=['국어','영어'])

# 전체컬럼의 슬라이싱 활용하여 특정컬럼 삭제할수있음
df.columns[3:5]
df.drop(columns=['국어','영어'])
df.drop(columns=df.columns[3:5])
```

```python
# drop 삭제 - index(row 값 삭제) : inplace=True
df.drop(index='1번')
df.head(1)
```
