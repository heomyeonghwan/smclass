```python
import matplotlib.pyplot as plt
#한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #win 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스
```

```python
#x,y축 데이터
x = [1,2,3]  #리스트,튜플,DataFrame 데이터 - 리스트형태,series
y= [2,4,8]
yy= [1,8,3]

#선그래프 생성
#linewidth 선두께 markeredgecolor 테두리선
plt.plot(x,y, label = '인원변동',color='pink',linewidth=5,marker='o',markersize=10,markeredgecolor='red') 
# marker 지점표시 o,x,v  alpha 투명도 설정  ls: 라인스타일 : 점선 --  dash 선 -. , -.,
# plt.plot(x,yy, label = '비용변동',marker='v',markersize=10,markerfacecolor='yellow',alpha=0.2)
# plt.plot(x,yy, label = '비용변동',marker='v',ms=10,mfc='green',alpha=0.4,ls='-.',mec='red')

## 설정별 단축표시 'bv:'
# b:선색상, v:마커표시, : 선모양
plt.plot(x,yy,'bv:', label = '비용변동')
# 상단타이틀
plt.title("라인그래프 LINE",size =30) # 글자크기

# ## 범례 : f그래프 선 설명문구, loc: 범례위치 upper,center, lower,left,center,right
# plt.legend(loc='upper center')
plt.legend(loc=(0.1,0.8)) #좌표값으로 위치 설정가능
```
