```python
import matplotlib.pyplot as plt
#한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #win 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스
```

```python
#x,y축 데이터
x = [1,2,3]  #리스트,튜플,DataFrame 데이터 - 리스트형태,series
y= [2,4,8]

plt.figure(figsize=(7,5))
plt.plot(x,y, label='범례데이터',marker='o')
## text(x좌표위치,y좌표위치,데이터값)
plt.text(x[0],y[0],y[0],ha='center')
plt.text(x[1],y[1],y[1],ha='center')
plt.title("라인그래프 LINE")
plt.legend()

plt.show()
```
