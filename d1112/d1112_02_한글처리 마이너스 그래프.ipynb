```python
import matplotlib.pyplot as plt
#한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #win 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스

# matplotlib.rcParams['font.family'] = 'AppleGothic'   # mac 용
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
#x,y축 데이터
x = [1,2,3]  #리스트,튜플,DataFrame 데이터 - 리스트형태,series
y= [2,4,8]

#선그래프 생성
plt.plot(x,y)
# 상단타이틀
plt.title("라인그래프 LINE")
plt.show()
```
