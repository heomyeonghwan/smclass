```python
import pandas as pd
import random
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
## 컬럼 전체수정
df['요일'] = 0
```

```python
## 특정 셀부분을 수정
df.loc['1번','요일'] = 5
df.iloc[1,9] =2 
df
```

```python
df.shape
#랜덤으로 0~6까지 숫자 입력
for i in range(8):
# df.loc['1번','요일'] = random.randint(0,6)
  df.iloc[i,9] = random.randint(0,6)
df
```

```python
def func(x):
  if x == 0: return '일요일'
  elif x == 1: return '월요일'
  elif x == 2: return '화요일'
  elif x == 3: return '수요일'
  elif x == 4: return '목요일'
  elif x == 5: return '금요일'
  elif x == 6: return '토요일'
```

```python
df['요일'].apply(func)
```

```python
df['요일'] = df['요일'].apply(func)
df
```

```python
## 결과 컬럼 생성
df['결과'] = '불합격'
df
## 국어점수가 60점이상이면 합격으로 수정
filt =df['국어'] >= 60
df.loc[filt,'결과'] ='합격'
df
```

```python
## 학교 + 등학교
## 키 + 'cm'

## astype() 타입변경후 문자열 + 문자열
df['키'] = df['키'].astype(str)+'cm'
df
```

```python
df = pd.read_csv('drinks.csv')
df
```

```python
## 모든 술의 총량을 출력하시오
### total_servings 컬럼을 구하시오. beer+ spirit+wine
#  / 전체 알콜의 도수를 확인

df['total_servings'] = df['beer_servings'] +df['spirit_servings'] +df['wine_servings']
df
```

```python
df['continent'].unique()
```

```python
filt = df['continent'].isnull()
df.loc[filt,'total_servings'].sum()
```

```python
## continent: 대륙 nan값을 
df['continent'] = df['continent'].fillna('OT',inplace=True)
```

```python
## 대륙별 술합계 total_servings 를 구하시오
df.groupby('continent')['total_servings'].sum().sort_values()
```

```python
df.columns
```

```python
## 술소비량 대비 알콜비율이 어느정도 되는지 출력하시오
## 알콜 비율을 계산해서 alcohol_rate 컬럼추가 
##'total_litres_of_pure_alcohol' 알콜비율을 total_servings으로 나눈것
## 술소비량 대비 알콜비율이 높은 나라별 20개를 출력
df['alcohol_rate'] = df['total_litres_of_pure_alcohol'] / df['total_servings']

## 대륙별 알콜비율도 계산해서 출력
```
